# Fireworks RAG Demo

This notebook runs the sample RAG app in `app/rag.py` using:
- Chat model: `accounts/fireworks/models/gpt-oss-20b`
- Embedding model: `accounts/fireworks/models/qwen3-embedding-8b` (serverless)
- Documents: PDF files in `data/`

```python
import os
import getpass
from pathlib import Path

from dotenv import load_dotenv

load_dotenv()

if not os.environ.get("FIREWORKS_API_KEY"):
    os.environ["FIREWORKS_API_KEY"] = getpass.getpass("Enter your Fireworks API key: ")

os.environ.setdefault("FIREWORKS_CHAT_MODEL", "accounts/fireworks/models/gpt-oss-20b")
os.environ.setdefault("FIREWORKS_EMBEDDING_MODEL", "accounts/fireworks/models/qwen3-embedding-8b")
os.environ.setdefault("RAG_DATA_DIR", "data")

data_dir = Path(os.environ["RAG_DATA_DIR"])
pdfs = list(data_dir.glob("**/*.pdf"))
print(f"Using {len(pdfs)} PDF file(s) from {data_dir.resolve()}")
for pdf in pdfs:
    print(f"- {pdf.name}")
```
